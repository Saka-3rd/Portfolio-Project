```python
# Average Yield as a predictor of Price

sns.regplot(x="Average Yield(tonnes per heactare)", y="Price(dollars per tonne)", data=canada_canola)
plt.ylim(0,)
```

```python
from scipy import stats
```

```python
# calculating the Pearson Correlation Coefficient and P-value of 'Average Yield' and 'Price'.

pearson_coef, p_value = stats.pearsonr(canada_canola['Average Yield(tonnes per heactare)'], canada_canola['Price(dollars per tonne)'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Creat a linear regression object

lm = LinearRegression()
lm
```

```python
# Showing Average yield prediction of price

X = canada_canola[['Average Yield(tonnes per heactare)']]
Y = canada_canola['Price(dollars per tonne)']
```

```python
# Fit the model using Average Yield

lm.fit(X,Y)
```

```python
# Output a prediction

Yhat=lm.predict(X)
Yhat[0:5] 
```

```python
# Value of the intercept

lm.intercept_
```

```python
# Value of the slope

lm.coef_
```

```python
# The final estimated linear model we get? with the structure:

# Yhat = a + bX

# Plugging in the actual values we get:

# Price = 34.902831382188765 + 216.80155553(Average Yield)
```

```python
#Calculating the R-square

print('The R-square is: ', lm.score(X, Y))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(canada_canola['Price(dollars per tonne)'], Yhat)
print('The mean square error of price and predicted value is: ', mse)
```
